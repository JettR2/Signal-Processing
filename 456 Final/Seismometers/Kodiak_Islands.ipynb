```python


# NS
NS_Samples = NS_Data["# dataset: GeoCSV 2.0"]  # 1 col. Displacements
Number_of_NS_samples = len(NS_Samples)  # # of Samples
NS_XTime_Seconds = Number_of_NS_samples / Sampling_Interval
NS_XTime_minutes = NS_XTime_Seconds / 60
NS_XTime_Hours = NS_XTime_minutes / 60

NS_time = np.linspace(0, NS_XTime_Hours, Number_of_NS_samples)

plt.figure()
#plt.subplot(3, 1, 1)
plt.plot(NS_time, NS_Samples)
plt.title('Kodiak Islands Seismogram')
# plt.xlabel('Hours')
plt.ylabel('N-S')
plt.ylim([-4 * 10**7, 4 * 10**7])
plt.xlim([0, NS_XTime_Hours])
plt.grid(True)
plt.show()
```

```python
# EW
EW_Samples = EW_Data["# dataset: GeoCSV 2.0"]
Number_of_EW_samples = len(EW_Samples)
EW_XTime_Seconds = Number_of_EW_samples / Sampling_Interval
EW_XTime_minutes = EW_XTime_Seconds / 60
EW_XTime_Hours = EW_XTime_minutes / 60
EW_time = np.linspace(0, EW_XTime_Hours, Number_of_EW_samples)

#plt.subplot(3, 1, 2)
plt.figure()
plt.plot(EW_time, EW_Samples)
# plt.xlabel('Number of Samples')
plt.ylabel('E-W - Acceleration m/s^2')
plt.ylim([-4 * 10**7, 4 * 10**7])
plt.xlim([0, EW_XTime_Hours])
plt.grid(True)
plt.show()
```

```python
# Z direction
Z_Samples = Z_Data["# dataset: GeoCSV 2.0"]
Number_of_Z_samples = len(Z_Samples)
Z_XTime_Seconds = Number_of_Z_samples / Sampling_Interval
Z_XTime_minutes = Z_XTime_Seconds / 60
Z_XTime_Hours = Z_XTime_minutes / 60
Z_time = np.linspace(0, Z_XTime_Hours, Number_of_Z_samples)

#plt.subplot(3, 1, 3)
plt.figure()
plt.plot(Z_time, Z_Samples)
plt.xlabel('Hours')
plt.ylabel('Z')
plt.ylim([-4 * 10**7, 4 * 10**7])
plt.xlim([0, Z_XTime_Hours])
plt.grid(True)

plt.show()
```
